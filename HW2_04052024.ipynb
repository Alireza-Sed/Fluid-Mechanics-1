```python
import numpy as np
import scipy as sp
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

### Assuming that Stokes'law applies:
#### du/dt + b*u = a ;(the predictive model of the velocity distribution for a falling particle)

```python
def dudt(t,u):
    # units of all variables and parameters are in SI.
    # gravitational acceleration
    g = 9.8 
    # fluid Viscosity 
    mu = 0.3
    # sphere density
    rho_s = 2700
    # fluid denisty
    rho_f = 678
    # Diameter
    D  = 4e-3 
    b= -18*mu/rho_s/D/D
    a= g*(1-rho_f/rho_s)
    return b*u + a

# initial velocity: falling from rest
u0 = 0
```

```python
t = np.linspace(0,0.1,100)
sol_u = odeint(dudt, y0 = u0, t=t, tfirst= True)
```

```python
velocity = sol_u.T[0]
plt.plot(t,velocity)
plt.title('Distribution Velocity of a falling particle in a fluid')
plt.ylabel('Velocity [m/s]')
plt.xlabel('Time [s]')
```

```python
u_terminal = np.round(velocity[-1],4)
print('Terminal Velocity [m/s] = ', u_terminal)
```
